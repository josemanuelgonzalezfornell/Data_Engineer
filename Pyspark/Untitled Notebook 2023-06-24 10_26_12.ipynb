Creamos un Dataframe

```python
columns = ["Age", "Name"]
data = [(33, "john jones"),
        (24, "tracey smith"),
        (40, "amy sanders")]

df_1 = spark.createDataFrame(data=data, schema=columns)

df_1.show(truncate=False) #Truncate False sirve para evitar que te acorte los nombres si son muy largos
```

```python
def upperCase(str):
    str = str + " ***"
    return str.upper()

upperCaseUDF = udf(lambda x: upperCase(x)) #udf sirve para iterar por columnas
```

```python
df_1.withColumn("Created Name", upperCaseUDF(df_1.Name)).show(truncate=False)

#la función withColumn crea una nueva columna con la query que hacemos (primero se pone el nombre, despues se pone la acción)
```

```python
from pyspark.sql.types import *
```

```python
data_schema = [StructField('name',StringType(),True), #El true es para decirle que acceda a esa columna
              StructField('email',StringType(),True),
              StructField('city',StringType(),True),
              StructField('mac',StringType(),True),
              StructField('timestamp',DateType(),True),
              StructField('creditcard',StringType(),True)]
```

```python
final_struc = StructType(fields = data_schema)
```

```python
people = spark.read.json("data/people.json", schema =final_struc)
```
